```python
df = data[['Pclass', 'Fare', 'Age', 'Sex', 'Survived']]
df = df.dropna(axis=0)
convert_sex = {'female':0, 'male':1}
df['Sex'] = df['Sex'].map(convert_sex)
df_base = df[['Pclass', 'Fare', 'Age', 'Sex']]
df_target = df[['Survived']]
df.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier(random_state=241)
clf.fit(df_base, df_target)
```

```python
importances = clf.feature_importances_
```
